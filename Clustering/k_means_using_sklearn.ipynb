```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(centers=3, n_samples=500, n_features=2, shuffle=True, random_state=40)
print(X.shape)
```

```python
clusters = len(np.unique(y))
print(clusters)
```

```python
k = KMeans(n_clusters=clusters, max_iter=150)
k.fit(X)
y_pred = k.predict(X)
```

```python
import matplotlib.pyplot as plt

# Extract the cluster centers
cluster_centers = k.cluster_centers_

# Scatter plot the data points with different colors for each cluster
for i in range(clusters):
    plt.scatter(X[y_pred == i, 0], X[y_pred == i, 1], label=f'Cluster {i + 1}')

# Plot the cluster centers as well
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='black', s=100, marker='x', label='Cluster Centers')

# Add labels and a legend
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.legend()
plt.show()
```
